for continuous features, if were to be categorical then bar chart!!

> bin size increase =bins widder, vice verse

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Data/mpg.csv")
df.head()
```

```python
df.columns
```

```python
data = [go.Histogram(
    x=df['mpg'],
    xbins=dict(start=8,end=50,size=6),
)]

layout=go.Layout(title="HISTOGRAM!!")

fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="Histogram.html")
```
