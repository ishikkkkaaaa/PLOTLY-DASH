```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
import numpy as np
```

> Objective: Create a histogram that plots the 'length' field <br>
> from the Abalone dataset (../data/abalone.csv). <br>
> Set the range from 0 to 1, with a bin size of 0.02

```python
df=pd.read_csv("../Data/abalone.csv")
df.head()
```

```python
df.columns
```

```python
data=[go.Histogram(x=df['length'],
                  xbins=dict(start=0,end=1,size=0.02))]
layout=go.Layout(title="HISTOGRAM!!")

fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="SolutionHistogram.html")
```
